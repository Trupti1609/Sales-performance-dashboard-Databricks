```python
%sql
use catalog sales;
use schema default ;
Select count(*) 
from sales.default.messy_sales_dataset_eda
```

```python
%sql
select count(distinct Order_ID  )
from messy_sales_dataset_eda
```

```python
%sql
SELECT 
count(*) - count(distinct Order_ID) as duplicate_rows
FROM messy_sales_dataset_eda
```

```python
%sql
create or replace temp view  Sales_NOduplicates as
select *
from
(select *, row_number ()over(partition by Order_Id order by Order_ID)as rn 
from messy_sales_dataset_eda)
where rn=1 ;
select count(*)
from Sales_NOduplicates
```

```python
%sql
select * from messy_sales_dataset_eda
```

```python
%sql
create or replace temp view Sales_valid  as 
select * 
from Sales_NOduplicates;

select *
from Sales_valid;
```

```python
%sql
create or replace temp view Discount_fix as 
select * ,CASE
           when Discount_Percent < 0 THEN 0
           ELSE Discount_Percent
       END AS Discount_Clean
FROM Sales_valid;
```

```python
%sql
select *
from Discount_fix
```

```python
%sql
    
create or replace temp view Sales_date_clean as 
select *,
       to_date(Order_Date,'dd-MM-yyyy') as Order_Date_clean
from Discount_fix
```

```python
%sql
select * from Sales_date_clean
```

```python
%sql
create or replace temp view Sales_final as 
select *,
 Unit_Price*Quantity*(1-Discount_Clean/100) as Revenue,
 Quantity * Cost_Price AS Total_Cost,
       (Quantity * Unit_Price * (1 - Discount_Clean/100)) 
       - (Quantity * Cost_Price) AS Profit
FROM sales_date_clean;
```

```python
%sql
select * 
from Sales_final
```

```python
%sql
select 
min(Unit_Price) as min_unit_price,
max(Unit_Price) as max_unit_price,
avg(Unit_Price) as avg_Unit_price,
min(Quantity) as Min_Quantity,
max(Quantity) as min_Quatity,
avg(Quantity) as Avg_Quantity
from messy_sales_dataset_eda
```

```python
------ Count missing values for key columns in messy_sales_dataset_eda
SELECT
    SUM(CASE WHEN Quantity IS NULL THEN 1 ELSE 0 END) AS missing_quantity,      -- Missing Quantity count
    SUM(CASE WHEN Unit_Price IS NULL THEN 1 ELSE 0 END) AS missing_price,       -- Missing Unit_Price count
    SUM(CASE WHEN Discount_Percent IS NULL THEN 1 ELSE 0 END) AS missing_discount, -- Missing Discount_Percent count
    SUM(CASE WHEN Region IS NULL THEN 1 ELSE 0 END) AS missing_region,          -- Missing Region count
    SUM(CASE WHEN Category IS NULL THEN 1 ELSE 0 END) AS missing_category       -- Missing Category count
FROM messy_sales_dataset_eda;
```

```python
%sql
-- Returns Order_IDs with duplicate entries and their counts
select Order_ID, count(*) as order_count
from messy_sales_dataset_eda
group by Order_ID
having count(*) > 1
```

```python
%sql
Select 
sum(Case WHEN Quantity <0 THEN 1 ELSE 0 END) as Invalid_Qunatity,
sum(Case WHEN Unit_Price <=0 THEN 1 ELSE 0 END) as Invalid_Price,
sum(Case WHEN Discount_Percent <0 THEN 1 ELSE 0 END) as Invalid_Discount
from messy_sales_dataset_eda
```

```python
%sql

CREATE OR REPLACE TEMP VIEW clean_no_duplicates AS
SELECT *
FROM (
    SELECT *,
           ROW_NUMBER() OVER (PARTITION BY Order_ID ORDER BY Order_ID) AS rn
    FROM sales.default.messy_sales_dataset_eda
) t
WHERE rn = 1;
```

```python
%sql
select count(*)
from clean_no_duplicates
```

```python
%sql
create or replace temp view sales_clean  as
select Order_ID,
to_date(Order_Date ,'dd-MM-yyyy') as Order_Date,
    INITCAP(TRIM(Region)) AS Region,
    INITCAP(TRIM(Category)) AS Category,
    CASE WHEN Quantity <= 0 THEN 0 ELSE Quantity END AS Quantity,
    CASE WHEN Unit_Price <= 0 THEN 0 ELSE Unit_Price END AS Unit_Price,
    CASE WHEN Discount_Percent < 0 THEN 0 ELSE Discount_Percent END AS Discount_Percent,
    Cost_Price,
    INITCAP(TRIM(Payment_Mode)) AS Payment_Mode,
    INITCAP(TRIM(Ship_Mode)) AS Ship_Mode
FROM clean_no_duplicates;
```

```python
%sql
create or replace table sales_clean_final as 
select *
from sales_clean
where Quantity is not null
and Unit_Price is not null
```

```python
%sql
select * from sales_clean_final
```

```python
%sql
CREATE OR REPLACE TABLE sales_gold AS
SELECT *,
       Quantity * Unit_Price * (1 - Discount_Percent/100) AS Revenue,
       Quantity * Cost_Price AS Total_Cost,
       (Quantity * Unit_Price * (1 - Discount_Percent/100)) 
           - (Quantity * Cost_Price) AS Profit,
       YEAR(Order_Date) AS Year,
       MONTH(Order_Date) AS Month,
       DATE_FORMAT(Order_Date, 'yyyy-MM') AS Year_Month
FROM sales_clean_final;
```

```python
%sql
select * from sales_clean_final
```
